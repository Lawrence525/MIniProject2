```python
import requests as re
import os
import json
from IPython.display import JSON
import pandas as pd
from pandas import json_normalize
from pandas.io.json import json_normalize
client_id = os.environ["YELPID"]
client_secret = os.environ["YELPKey"]
```

```python
headers = {'Authorization': 'Bearer %s' % client_secret}
```

```python
url='https://api.yelp.com/v3/businesses/search'

params = {'term':'sushi','location':'Windsor,ON'}
res = re.get(url, params=params,headers=headers)
```

```python
data=res.json()
```

```python
JSON(data)
```

```python
df = pd.json_normalize(data,'businesses')
```

```python
df
```

```python
newdf=df[["location.city","rating","name","review_count"]]#narrow my POI
```

```python
newdf
```

```python
newdf.sort_values(by = ['location.city','rating','review_count'], ascending=(False,False,False))# avoid go to state now
```

```python
import sqlite3
database = "example.sqlite"          #save to SQLite
conn = sqlite3.connect(database)
```

```python
newdf.to_sql(name='data3', con=conn,if_exists='replace')
conn.close()
```
